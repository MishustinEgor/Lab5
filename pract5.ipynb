```python
import pandas as pd
from datetime import datetime
class FlightManager:
    def __init__(self):
        self.columns = ['destination', 'flight_number', 'aircraft_type']
        self.df = pd.DataFrame(columns=self.columns)
    def add_flight(self, destination, flight_number, aircraft_type):
        new_data = {
            'destination': destination,
            'flight_number': flight_number,
            'aircraft_type': aircraft_type
        }
        self.df = pd.concat([self.df, pd.DataFrame([new_data])], ignore_index=True)
        self.df.sort_values(by='destination', inplace=True)
    def search_flights(self, aircraft_type):
        result = self.df[self.df['aircraft_type'] == aircraft_type]
        return result[['destination', 'flight_number']] if not result.empty else None
    def delete_by_column(self, column_name, value):
        if column_name not in self.columns:
            return 0
        before = len(self.df)
        self.df = self.df[self.df[column_name] != value]
        return before - len(self.df)
    def save_to_parquet(self, filename):
        self.df.to_parquet(filename, engine='pyarrow')
    def load_from_parquet(self, filename):
        try:
            self.df = pd.read_parquet(filename, engine='pyarrow')
        except:
            self.df = pd.DataFrame(columns=self.columns)
manager = FlightManager()
manager.load_from_parquet('flights.parquet')
while True:
    print("\nМеню:")
    print("1. Добавить рейс")
    print("2. Найти рейсы по типу самолета")
    print("3. Удалить рейсы по номеру")
    print("4. Показать все рейсы")
    print("5. Выход")
    choice = input("Выберите действие: ")
    if choice == '1':
        destination = input("Введите пункт назначения: ")
        flight_number = input("Введите номер рейса: ")
        aircraft_type = input("Введите тип самолета: ")
        manager.add_flight(destination, flight_number, aircraft_type)
        print("Рейс успешно добавлен!")
    elif choice == '2':
        aircraft_type = input("Введите тип самолета для поиска: ")
        result = manager.search_flights(aircraft_type)
        if result is None:
            print("Рейсов с таким типом самолета не найдено.")
        else:
            print("\nНайденные рейсы:")
            for _, row in result.iterrows():
                print(f"Пункт: {row['destination']}, Рейс: {row['flight_number']}")
    elif choice == '3':
        flight_number = input("Введите номер рейса для удаления: ")
        deleted = manager.delete_by_column('flight_number', flight_number)
        print(f"Удалено рейсов: {deleted}")
    elif choice == '4':
        print("\nВсе рейсы:")
        if manager.df.empty:
            print("Нет данных о рейсах")
        else:
            for _, row in manager.df.iterrows():
                print(f"Пункт: {row['destination']}, Рейс: {row['flight_number']}, Самолет: {row['aircraft_type']}")
    elif choice == '5':
        manager.save_to_parquet('flights.parquet')
        print("Данные сохранены. Выход из программы.")
        break
    else:
        print("Неверный ввод. Попробуйте снова.")
```
